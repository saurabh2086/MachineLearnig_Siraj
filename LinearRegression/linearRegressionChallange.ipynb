# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("ggplot")
```

# Loading Data

```python
dataframe = pd.read_csv('challenge_dataset.txt', header=None, delimiter=",")
dataframe.columns =['x','y']
x_values = dataframe[['x']]
y_values = dataframe[['y']]
```

```python
dataframe.head()
```

# Trainig Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_values, y_values)
```

# Visalize results 

```python
plt.scatter(x_values, y_values)
plt.plot(x_values, model.predict(x_values))
plt.show()
```

# Plotting Error

```python
plt.scatter(x_values, (y_values - model.predict(x_values))**2)
plt.hlines(y=0,xmax=25, xmin=0, colors="red")
plt.show()
```
